# Finding optimal number of clusters

```python
# find the optimal number of clusters
# It can be find by two ways
# 1.Elbow method
# 2.silhoutte method
```

# Elbow method

```python
wcss=[]
for i in np.arange(1,11):
    km_cl=KMeans(n_clusters=i,random_state=1)
    km_cl.fit(continous_var)
    wcss.append(km_cl.inertia_)
```

```python
# plotting between number of clusters and wcss
plt.plot(range(1,11),wcss)
plt.title('Elbow plot')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
```

# silhoutte score method

```python
sil_score=[]
for i in np.arange(2,11):
    km_cl_sil=KMeans(n_clusters=i,random_state=1)
    km_cl_sil.fit(continous_var)
    km_pre=km_cl_sil.fit_predict(continous_var)
    print('silhoute score for',i,'cluster is',silhouette_score(continous_var,km_pre))
    sil_score.append(silhouette_score(continous_var,km_pre))
```

```python
# plotting bar plot between number of clsuters and silhoutte scores
sns.barplot(np.arange(2,11),sil_score)
```

```python
# From the above elbow plot and silhoutte score method we can say that optimal number of cluster is 2.
# Again it can be cross checked with silhoutte visualizer
```

# Silhoutte visualizer

```python
km_sil_viz=KMeans(n_clusters=2,random_state=1)
km_sil_viz=SilhouetteVisualizer(km_sil_viz)
km_sil_viz.fit(continous_var)
km_sil_viz.show()
```

```python
# there are two conditions for silhoutte visualizer.
# 1.there should be no or little outliers
# 2.All the clusters should cross the average silhoutte score
# From the above plot we can confirm that optimal number of clusters i.e 2, satisfied both the conditions
# Hence optimal number of clsuter is 2
```
